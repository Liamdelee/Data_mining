<h1>Number of inhabitants in Belgium by nationality</h1>

```python
test1 = []
test2 = []

migratiedata = pd.read_excel('journalistiekdata.xls', sheetname='test1.0')
landen = pd.read_excel('countries-name.xls', sheetname='countries-20140629')
landen = landen['Nederlands naam']

BelTotIndex = migratiedata.query('NISCODE == 1000').index[0]

for x in range(0,93):
    test1.append(migratiedata.iloc[(BelTotIndex+2)][landen[x]])

for y in range(0,93):
    test2.append(landen.iloc[y])
    
df = pd.DataFrame()
df['Landen'] = test2
df['Totaal'] = test1

#definition of the boundaries in the map
district_geo = 'countries.geo.json'
    
#creation of the choropleth
map1 = folium.Map(location=[0, 0], zoom_start=1.5)
map1.geo_json(geo_path = district_geo,
              data = df,
              columns = ['Landen', 'Totaal'],
              key_on = 'feature.properties.name',
              threshold_scale=[300, 3000, 7000, 12000, 25000, 40000],
              fill_color = 'YlOrRd',
              fill_opacity = 0.7,
              line_opacity = 0.2,
              legend_name = 'Number of inhabitants by nationality in Belgium' )
map1
```

<h1>Requests for Asylum België 2012</h1>

```python
asielaanvragen = pd.read_excel('asielaanvragen2.0.xlsx', sheetname='page 7')

asielaanvragen.columns.values[1] = 'twee'
asielaanvragen.columns.values[2] = 'drie'
asielaanvragen.columns.values[3] = 'vier'
asielaanvragen.columns.values[4] = 'vijf'

#definition of the boundaries in the map
district_geo = 'countries.geo.json'

for x in('twee', 'drie', 'vier', 'vijf'):
    asielaanvragen[x].fillna(0, inplace=True)


#creation of the choropleth
map2 = folium.Map(location=[0, 0], zoom_start=1.5)
map2.geo_json(geo_path = district_geo,
    data = asielaanvragen,
    columns = ['Land van herkomst', 'twee'],
    threshold_scale=[50, 100, 500, 1000, 2500],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrRd',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = 'Request for Asylum Belgium 2012' )
map2
```
